```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
#HISTOGRAMA

os.chdir('../01b_Analisis de datos//')
datos = pd.read_csv('AUTOS.csv')
plt.figure()
plt.hist(datos[' curb-weight'], bins = 10, edgecolor = 'black', \
         linewidth=1)
plt.xlabel('curb-weight')
plt.ylabel('Frecuencia')
```

```python
#DIAGRAMA DE CAJAS

os.chdir('../01b_Analisis de datos//')
datos = pd.read_csv('AUTOS.csv')
plt.figure()
datos.boxplot()
plt.figure()
datos.boxplot(column=[' height', ' width'], grid=False, vert=False)
```

```python
#DIAGRAMA DE BARRAS

os.chdir('../01b_Analisis de datos//')
datos = pd.read_csv('AUTOS.csv')
barras = pd.value_counts(datos[' height'])
plt.figure()
N=len(barras)
plt.bar(np.arange(N), barras)  # Gráfico de barras
plt.title(' Height')      # Colocamos el título
plt.ylabel('Frecuencia')
# Colocamos las etiquetas del eje x
plt.xticks(np.arange(N), barras.index)
```

```python
#DIAGRAMA DISPERSION

os.chdir('../01b_Analisis de datos//')
datos = pd.read_csv('AUTOS.csv')

plt.figure()
plt.scatter(datos[' height'],datos[' width'], marker='*')
plt.ylabel(' width')
plt.xlabel(' height')

conlenght = datos.query(" make== ' honda'")
otracaracteristica = datos.query(" make != ' honda'")

fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.6, 0.8])

p1 = plt.scatter(conlenght[' height'], conlenght[' width'],  marker='^', label=' honda')
p2 = plt.scatter(otracaracteristica[' height'], otracaracteristica[' width'],  marker='o', label='Otra')
plt.legend(handles=[p1, p2], bbox_to_anchor=(1.05, 1), loc='upper left')
plt.ylabel('width')
plt.xlabel('height') 
```
