```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
datos = pd.read_csv('../01b_Analisis de datos/AUTOS.csv')
#--Matriz de correlacion --
print(datos.corr())

sns.heatmap(datos.corr(), square=True, annot=True)

plt.figure()
plt.scatter(datos[' horsepower'],datos[' city-mpg'])
plt.ylabel(' city-mpg')
plt.xlabel(' horsepower')
```
