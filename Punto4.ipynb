```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
os.chdir('../01b_Analisis de datos//')
datos = pd.read_csv('AUTOS.csv')
datos.boxplot()
plt.figure()
datos.boxplot(column=[' horsepower'], grid=False, vert=False)
```

```python
datos[' horsepower']  # as a Series
horsepower= datos[' horsepower'].tolist()  # as a numpy list
newhorsepower = [x for x in horsepower if np.isnan(x) == False]
print('El Q1 es: ', np.percentile(newhorsepower, 25)) #primer cuartil
print('El Q2 es: ', np.percentile(newhorsepower, 50)) #segundo cuartil
print('El Q3 es: ', np.percentile(newhorsepower, 75)) #tercer cuartil
print('El Q4 es: ', np.percentile(newhorsepower, 100)) #cuarto cuartil

#rango intercuartil
def IQR(newhorsepower):
    return np.percentile(newhorsepower, 75) - np.percentile(newhorsepower, 25)
print ('El RIC es: ', IQR(newhorsepower))
```
